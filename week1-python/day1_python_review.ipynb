```python
# 1. 变量和基本类型
name = "周婧怡"
age = 19
gpa = 3.7
print(f"我叫{name}，今年{age}岁，绩点{gpa}")
```

```python
# 2. 列表（类似C的数组，但更强大）
scores = [91, 85, 88, 92, 79]
scores.append(95)  # 添加元素
print("成绩列表:", scores)
print("最高分:", max(scores))
print("平均分:", sum(scores) / len(scores))

# 列表推导式（Python特色，必会！）
squared = [x**2 for x in range(10)]
print("0-9的平方:", squared)
```

```python
# 3. 字典（C语言没有，非常重要）
student = {
    "name": "周婧怡",
    "major": "人工智能",
    "courses": ["高数", "线代", "C语言"]
}
print("姓名:", student["name"])
print("专业:", student["major"])
student["grade"] = 2024  # 添加新键值对
print("完整信息:", student)
```

```python
# 4. if/else 和循环
score = 91
if score >= 90:
    print("优秀")
elif score >= 80:
    print("良好")
else:
    print("继续努力")

# for 循环
for i, s in enumerate(scores):
    print(f"第{i+1}门课: {s}分")
```

```python
# 5. 函数定义
def calculate_bmi(weight, height):
    """计算BMI：体重(kg) / 身高(m)的平方"""
    return weight / (height ** 2)

my_bmi = calculate_bmi(70, 1.75)
print(f"BMI: {my_bmi:.1f}")
```
